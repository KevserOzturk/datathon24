```python
# Gerekli kütüphaneleri yükle
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# CSV dosyalarının yolunu belirt
new_test_df_a_path = '/content/drive/My Drive/datathon/new_test_df.csv'
new_train_df_a_path = '/content/drive/My Drive/datathon/new_train_df.csv'


# CSV dosyalarını pandas ile oku
new_test_df_a = pd.read_csv(new_test_df_a_path)
new_train_df_a = pd.read_csv(new_train_df_a_path)
```

```python
# Sütunlardaki benzersiz veri tiplerini ve NaN değer sayısını bastırmak için kod

# Her sütunun veri tiplerini ve NaN sayılarını bulmak
for column in new_test_df_a.columns:
    unique_values = new_test_df_a[column].unique()  # Benzersiz değerler
    nan_count = new_test_df_a[column].isnull().sum()  # NaN sayısı

    # Benzersiz değerleri ve NaN sayısını yazdır
    print(f"Sütun: {column}")
    print(f"Benzersiz Değerler: {unique_values}")
    print(f"NaN Değer Sayısı: {nan_count}")
    print("-" * 50)
```

```python
# Belirtilen sütunları veri setinden çıkarma
columns_to_drop = ['Girisimcilikle Ilgili Deneyiminizi Aciklayabilir misiniz?',
                   'Hangi STK\'nin Uyesisiniz?',
                   'Lise Adi Diger']

# train ve test veri setlerinden bu sütunları çıkarma
new_test_df_a = new_test_df_a.drop(columns=columns_to_drop)

# Sonuçları kontrol etmek için
print("Train DataFrame - First 5 Rows after dropping columns:")
print(new_test_df_a.head())
```

```python
import pandas as pd

# Dogum Tarihi sütununuzdaki verileri otomatik olarak tarih formatına çevirme
new_test_df_a['Dogum Tarihi'] = pd.to_datetime(new_test_df_a['Dogum Tarihi'], errors='coerce')

# Tarih kısmını YYYYMMDD formatında int haline çevirme
new_test_df_a['Dogum Tarihi'] = new_test_df_a['Dogum Tarihi'].dt.strftime('%Y%m%d').astype(float)

# Sonuçları kontrol etme
print(new_test_df_a[['Dogum Tarihi']].head())
```

```python
import pandas as pd
import xgboost as xgb

# Modeli yükleme
model = xgb.Booster()
model.load_model('/content/drive/My Drive/datathon/xgboost_model.json')
```

```python
from sklearn.preprocessing import LabelEncoder

# Kategorik sütunları belirleme
cat_features = ['Dogum Yeri', 'Ikametgah Sehri', 'Universite Adi', 'Bölüm',
                'Lise Adi', 'Lise Sehir', 'Lise Turu', 'Lise Bolumu',
                'Burs Aldigi Baska Kurum', 'Baska Kurumdan Aldigi Burs Miktari']

# Her kategorik sütunu dönüştürme
le = LabelEncoder()
for col in cat_features:
    new_test_df_a[col] = le.fit_transform(new_test_df_a[col].astype(str))

# NaN değerleri -1 ile doldurma
new_test_df_a.fillna(-1, inplace=True)

# Yeniden DMatrix formatına çevirme ve tahmin yapma
X_test = xgb.DMatrix(new_test_df_a.drop(columns=['id']))
y_pred = model.predict(X_test)

# Sonuçları kaydetme
results_df = pd.DataFrame({
    'id': new_test_df_a['id'],
    'Degerlendirme Puani': y_pred
})

results_df.to_csv('tahmin_sonuclari_test3.csv', index=False)
print("Tahmin sonuçları kaydedildi.")
```
