```python
# Gerekli kütüphaneleri yükle
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# CSV dosyalarının yolunu belirt
test_encode_path = '/content/drive/My Drive/datathon/test_df_encode_edilmis.csv'
new_train_df_path = '/content/drive/My Drive/datathon/new_train_df.csv'


# CSV dosyalarını pandas ile oku
test_df_encode = pd.read_csv(test_encode_path)
new_train_df_t = pd.read_csv(new_train_df_path)
```

```python
# DataFrame'leri görüntüle (ilk 10 satır)
print("Test Encode DataFrame - First 10 Rows:")
display(test_df_encode.head(10))  # İlk 10 satırı göster
```

```python
# DataFrame'deki sütun isimlerini ve her sütundaki veri sayısını göstermek için kod
column_info = test_df_encode.count()

# Sütun isimleri ve veri sayısını yazdır
for column, count in column_info.items():
    print(f"Sütun: {column}, Veri Sayısı: {count}")
```

```python
# Tüm string verileri küçük harfe çevirme ve eksik değerleri -1 ile değiştirme
test_df_encode = test_df_encode.applymap(lambda s: s.lower() if isinstance(s, str) else s)

# Kontrol etme
print(test_df_encode.head())
```

✅

```python
# Tüm sütunların veri tiplerini görmek
print(test_df_encode.dtypes)
```

```python
# Tüm sütunların veri tiplerini görmek
print(new_train_df_t.dtypes)
```

```python
# Sütun veri tiplerini karşılaştırma
for column in new_train_df_t.columns:
    if column in test_df_encode.columns:
        train_dtype = new_train_df_t[column].dtype
        test_dtype = test_df_encode[column].dtype

        # Eğer veri tipleri farklıysa ekrana bastır
        if train_dtype != test_dtype:
            print(f"Sütun: {column} -> new_train_df_t dtype: {train_dtype}, test_df_encode dtype: {test_dtype}")
```

```python
# new_train_df_t'deki sütunlara göre test_df_encode'deki sütunları dönüştürme
for column in new_train_df_t.columns:
    if column in test_df_encode.columns:
        train_dtype = new_train_df_t[column].dtype
        test_dtype = test_df_encode[column].dtype

        # Eğer veri tipi farklıysa, test_df_encode sütununu new_train_df_t veri tipiyle dönüştür
        if train_dtype != test_dtype:
            try:
                test_df_encode[column] = test_df_encode[column].astype(train_dtype)
                print(f"{column} sütunu {train_dtype} tipine dönüştürüldü.")
            except ValueError as e:
                print(f"{column} sütunu dönüştürülürken hata oluştu: {e}")
```

✅

```python
# Tüm eksik değerleri (-1) ile değiştirme, NaN dahil
test_df_encode.fillna(-1, inplace=True)

# String olarak "nan", "NaN" gibi varyasyonları da -1 yapma
test_df_encode.replace(to_replace=['nan', 'NaN', 'NAN', 'Nan', 'NAn', 'nAn','-', None], value=-1, inplace=True)
```

```python
# DataFrame'deki NaN değerlerinin toplam sayısını sütunlar bazında göstermek
nan_values = test_df_encode.isnull().sum()

# Eğer herhangi bir sütunda NaN değeri varsa bunları yazdır
nan_columns = nan_values[nan_values > 0]

if not nan_columns.empty:
    print("Veri setindeki NaN değerleri bulunan sütunlar ve sayıları:")
    print(nan_columns)
else:
    print("Veri setinde hiç NaN değeri yok.")
```

```python
# Drop etmek istediğin sütunların listesini belirle
columns_to_drop = [
    'Anne Egitim Durumu',
    'Baba Egitim Durumu',
    'anne_egitim_yeni2',
    'baba_egitim_yeni2',
    'Lise Mezuniyet Notu ort',
    'Lise Mezuniyet Notu',
    'Universite Not Ortalamasi',
    'Universite Not Ortalamasi Detayli'
]

# Mevcut dataframe'den bu sütunları drop et
new_test_df = test_df_encode.drop(columns=columns_to_drop)

# Sonucu kontrol et (ilk 10 satır)
print("New DataFrame (first 10 rows):")
display(new_test_df.head(10))
```

```python
import pandas as pd
from catboost import CatBoostRegressor

# Modeli yükleme
model = CatBoostRegressor()
model.load_model('/content/drive/My Drive/datathon/catboost_model.cbm')

# Kategorik sütunları belirtin
cat_features = ['Dogum Tarihi', 'Dogum Yeri', 'Ikametgah Sehri', 'Universite Adi',
       'Bölüm', 'Lise Adi', 'Lise Sehir', 'Lise Turu', 'Lise Bolumu',
       'Burs Aldigi Baska Kurum', 'Baska Kurumdan Aldigi Burs Miktari',
       'Hangi STK\'nin Uyesisiniz?',
       'Girisimcilikle Ilgili Deneyiminizi Aciklayabilir misiniz?']


# Test setinde tahmin yapma (y bağımlı değişken değil, sadece X kullanıyoruz)
X_test = new_test_df.drop(columns=['id'])

# Yüklenen model ile tahmin yapma
y_pred_loaded_model = model.predict(X_test)

# Tahmin sonuçlarını id ile birleştir
results_test_df = pd.DataFrame({
    'id': new_test_df['id'],  # Test setindeki 'id' sütunu
    'Degerlendirme Puani': y_pred_loaded_model  # Tahmin edilen sonuçlar
})

# Sonuçları CSV formatında kaydetme (başlıklar dahil)
results_test_df.to_csv('tahmin_sonuclari_test2.csv', index=False)

print("Tahmin sonuçları 'tahmin_sonuclari_test.csv2' dosyasına başarıyla kaydedildi.")
```
