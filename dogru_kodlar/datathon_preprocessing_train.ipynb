**Bu verileri encoding işlemi yapmadan önce anlamlı hale getirmeniz gerekiyor, çünkü burada hem sayı aralıkları hem de metinsel ifadeler karışık halde. Öncelikle kategorilerde standartlaştırma yapmalısınız. Aşağıdaki adımları izleyerek encoding işlemini gerçekleştirebilirsiniz:

1. Veriyi Standartlaştırma (Normalization)
Öncelikle aynı anlamı taşıyan ancak farklı yazılan değerleri tek bir forma indirgemelisiniz.

'ORTALAMA BULUNMUYOR' ve 'Ortalama bulunmuyor' → 'Ortalama Yok'
'Not ortalaması yok' → 'Ortalama Yok'
'4.0-3.5' ve '4-3.5' → '3.50 - 4.00'
'3.50-3' ve '3.00 - 3.50' → '3.00 - 3.50'
'2.50 -3.00' ve '2.50 - 3.00' → '2.50 - 3.00'
Diğer sayısal aralıklar da benzer şekilde standart hale getirilmeli.
Bu işlemi yaptıktan sonra verileri encoding'e hazır hale getirebilirsiniz.



```python
# Daha ayrıntılı ortalama encoding yapmak için yeni bir sütun ekleme

# Universite Not Ortalamasi sütunundaki string verileri temizlemek (boşluklar ve büyük/küçük harf farkı)
train_df['Universite Not Ortalamasi'] = train_df['Universite Not Ortalamasi'].str.lower().str.strip()

# Boşlukları düzenleme (örneğin '3.50-4.00' -> '3.50 - 4.00' olacak şekilde)
train_df['Universite Not Ortalamasi'] = train_df['Universite Not Ortalamasi'].str.replace(r'\s*-\s*', ' - ', regex=True)

train_df['Universite Not Ortalamasi Yeni'] = train_df['Universite Not Ortalamasi'].replace({
    'ORTALAMA BULUNMUYOR': 'nan',
    'Ortalama bulunmuyor':'nan',
    'Not ortalaması yok': 'nan',
    'not ortalaması yok': 'nan',
    'Hazırlığım': 'nan',
    'hazırlığım': 'nan',
    '0 - 1.79': 0,
    '1.00 - 2.50': 1,  # Farklı bir kategoriye yakınsa ona dahil edilebilir
    '1.80 - 2.49': 1,
    '2.00 - 2.50' : 1,
    '2.50 ve altı': 1,
    '2.50 - 2.99': 2,
    '2.50 - 3.00': 2,
    '3.00 - 2.50': 2,  # Yanlış format düzeltme
    '3.00 - 3.49': 3,
    '3.50 - 3': 3,  # Farklı bir kategoriye yakınsa ona dahil edilebilir
    '3.00 - 3.50': 3,
    '3.00 - 4.00': 3,
    '3.50 - 4.00': 4,
    '4 - 3.5': 4,
    '4.0 - 3.5': 4,
})

# Eski encoding mevcut 'Universite Not Ortalamasi' sütununda kalır ve
# yeni detaylı encoding "Universite Not Ortalamasi Yeni" sütununda yer alır
```
